```python
folder_path = "D:\\HSE\\I WILL BE THE BEST IN ML\\NLP course project Spot the Bot\\Danish_eBooks_Non_Poetry"
```

```python
import os
import pandas as pd
from collections import Counter

def count_words(text):
    return len(text.split())

def count_paragraphs(text):
    return len([p for p in text.split("\n") if p.strip()])

def count_characters(text):
    return len(text)

def extract_metadata(lines):
    author, title = None, None
    for line in lines:
        if "author:" in line.lower():
            author = line.split(":", 1)[-1].strip()
        elif "title:" in line.lower():
            title = line.split(":", 1)[-1].strip()
        if author and title:
            break
    return author, title

data = []

for file_name in os.listdir(folder_path):
    if file_name.endswith(".txt.utf-8"):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, "r", encoding="utf-8") as file:
            text = file.read()
            lines = text.split("\n")
            author, title = extract_metadata(lines)
            word_count = count_words(text)
            paragraph_count = count_paragraphs(text)
            char_count = count_characters(text)
            data.append({
                "File Name": file_name,
                "Author": author,
                "Title": title,
                "Word Count": word_count,
                "Paragraph Count": paragraph_count,
                "Character Count": char_count
            })

df = pd.DataFrame(data)

max_words = df["Word Count"].max()
min_words = df["Word Count"].min()
avg_words = df["Word Count"].mean()

max_paragraphs = df["Paragraph Count"].max()
min_paragraphs = df["Paragraph Count"].min()
avg_paragraphs = df["Paragraph Count"].mean()

max_chars = df["Character Count"].max()
min_chars = df["Character Count"].min()
avg_chars = df["Character Count"].mean()

print("Word Count Statistics:")
print(f"Max: {max_words}, Min: {min_words}, Avg: {avg_words:.2f}")

print("Paragraph Count Statistics:")
print(f"Max: {max_paragraphs}, Min: {min_paragraphs}, Avg: {avg_paragraphs:.2f}")

print("Character Count Statistics:")
print(f"Max: {max_chars}, Min: {min_chars}, Avg: {avg_chars:.2f}")
```
